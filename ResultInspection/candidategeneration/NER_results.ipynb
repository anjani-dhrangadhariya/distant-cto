```python
plt.figure(figsize = (6,5))
labels = ['1', '0']
ax = plt.axes()
sn.set(font_scale=1.5) # for label size
sn.heatmap(np.array(cm_distant_conf1_per), annot=True, annot_kws={"size": 24}, linewidths=.5) # font size
#ax.set_title('Confusion matrix for Exp 1.1 (EBM-PICO benchmark)',fontsize=15)
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```
